<b> <h3> <center> KMeans Clustering using Iris Dataset

```python
import pandas as pd
import numpy as np
from sklearn import *
from sklearn import datasets
```

```python
df = datasets.load_iris()
```

```python
x = df.data[:, :2]
y = df.target
```

```python
#attributes = ["sepal_length", "sepal_width", "petal_length", "petal_width", "class"]
#df.columns = attributes
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import scale
model = KMeans(n_clusters=3, random_state=5)   # using k = 3 ie. 3 clusters
model.fit(x) #fitting the model

#output shows summary of our model
```

```python
newlabels = model.labels_   #predicted labels
```

<b> Visualizing the actual and the predicted labels:

```python
import matplotlib.pyplot as plt
color_theme = np.array(['darkgray','salmon','skyblue'])
plt.scatter(x= x[:, 0], y= x[:, 1], c= color_theme[y])   #actual labels
```

```python
import matplotlib.pyplot as plt
color_theme = np.array(['darkgray','salmon','skyblue'])
plt.scatter(x= x[:, 0], y= x[:, 1], c= color_theme[newlabels])   #predicted labels
```

<b> From the above two graphs, we see that the model did a pretty good job of classifying the irises.
